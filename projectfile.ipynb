#Majority of coustomer order from??

```python
sns.countplot(x = df['listed_in(type)'])
plt.xlabel('Resturent Type')
plt.title('Count of Resturent')
plt.show()
```

#Conclusion: Majority of coustomer order from dining

```python
df.head(5)
```

#How many votes recived of resturent from costomer?  

solve by using groupby

```python
total_votes = df.groupby('listed_in(type)')['votes'].sum()
```

```python
print(total_votes)
```

#Draw a line chart

```python
plt.plot(total_votes,marker = '*')
plt.xlabel('Resturent type')
plt.ylabel('votes')
plt.title('Total votes recived')
plt.show()
```

#Conclusion: dining returents recived the highest votes from costomer.

#Box and whisker- rating analysis by order mode

```python
sns.boxplot(x = 'online_order', y = 'rate', data = df)
plt.title('Rating analysis by order mode')
plt.show()
```

#Heat map - graphical representation of data that use colour to show different values.

```python
pivot_table = df.pivot_table(df,index = 'listed_in(type)',columns = 'online_order',aggfunc = 'size' , fill_value = 0)
sns.heatmap(pivot_table , cmap = 'YlGnBu',  annot=True, fmt="d")
plt.show()
```
